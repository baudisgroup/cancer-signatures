```python
def loadpickle(filepath):
    with open(filepath, 'rb') as fi:
        return pickle.load(fi)
```

```python
def plotHistory(his_list, labels, title, loss_y_top=0):
    plt.figure(figsize=(15,7))
    plt.suptitle(title)
    plt.subplot(1,2, 1)
    for his in his_list:
        plt.plot(his.history['loss'])
    plt.title('Models loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    if loss_y_top > 0:
        plt.ylim(top=loss_y_top)
    plt.legend(labels, loc='upper right')

    plt.subplot(1,2, 2)
    for his in his_list:
        plt.plot(his.history['acc'])
    plt.title('Models accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(labels, loc=4)

    plt.show()
    
def plotHistory1(his_list, labels, title, loss_y_top=0):
    plt.figure(figsize=(15,7))
    plt.suptitle(title)
    plt.subplot(1,2, 1)
    for his in his_list:
        plt.plot(his['loss'])
    plt.title('Models loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    if loss_y_top > 0:
        plt.ylim(top=loss_y_top)
    plt.legend(labels, loc='upper right')

    plt.subplot(1,2, 2)
    for his in his_list:
        plt.plot(his['acc'])
    plt.title('Models accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(labels, loc=4)

    plt.show()
```
